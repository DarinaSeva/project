```python
import pandas as pd
from transliterate import translit, get_available_language_codes

def is_latin(s):
    try:
        s.encode(encoding='utf-8').decode('ascii')
    except UnicodeDecodeError:
        return False
    else:
        return True


df_encoded.columns = [translit(name, 'ru', reversed=True) if not is_latin(name) else name for name in df_encoded.columns]
df_encoded.columns = df_encoded.columns.str.replace(' ', '_')

# Показать обновленные названия столбцов
print(df_encoded.columns)
```
